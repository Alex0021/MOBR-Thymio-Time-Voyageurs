```python
from tdmclient import ClientAsync, aw
client = ClientAsync()
node = await client.wait_for_node()
await node.lock()
client.process_waiting_messages()


import os
import sys
import numpy as np
import math
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

sys.path.insert(0, os.path.join(os.getcwd(), 'Vision'))
sys.path.insert(0, os.path.join(os.getcwd(), 'GlobalNav'))
sys.path.insert(0, os.path.join(os.getcwd(), 'LocalNav'))
sys.path.insert(0, os.path.join(os.getcwd(), 'Kalman'))
sys.path.insert(0, os.path.join(os.getcwd(), 'sensors'))
sys.path.insert(0, os.path.join(os.getcwd(), 'plot'))
sys.path.insert(0, os.path.join(os.getcwd(), 'NMPC'))

from local_occupancy import thymio_coords
from plot import plot
from nmpc import nmpc

#from Vision import Vision
from LocalNav import localNav
#from GlobalNav import GlobalNav
#from Kalman import Kalman

```

```python
# Position #math.pi/2
abs_pos = [[50, 50, math.pi/2]] # absolute x,y,theta coordinates of the robot
# Goal Position
goal_position = [[80, 80]]

# Create Global Map
def create_map_global():
    # Create a xy map with all zeros
    map_size_x = 100
    map_size_y = 100
    map_global = np.zeros((map_size_x, map_size_y))

    # Set the edges to 1
    map_global[0, :] = 1  # Top edge
    map_global[-1, :] = 1  # Bottom edge
    map_global[:, 0] = 1  # Left edge
    map_global[:, -1] = 1  # Right edge
    
    return map_global

```

```python
map_global = create_map_global()

controller_r = 0
controller_l = 0

# Main loop
#i=0
while True:
    # Update Sensors
    #aw(node.lock())
    await node.wait_for_variables({"prox.horizontal"})
    prox_horizontal = list(node.v.prox.horizontal)
    
    # Update State
    # abs_pos = Vision()
    d = 9.5

    abs_pos[0][0] = abs_pos[0][0] + ((controller_r + controller_l) / 2) * math.cos(abs_pos[0][2])# * math.cos(abs_pos[0][2]+theta)
    abs_pos[0][1] = abs_pos[0][1] + ((controller_r + controller_l) / 2) * math.sin(abs_pos[0][2])# * math.sin(abs_pos[0][2]+theta)
    abs_pos[0][2] = abs_pos[0][2] - ((controller_r - controller_l) / (2 * d))
    #print(i)
    #print(abs_pos)
    # Kalman Filter
    # abs_pos = Kalman(abs_pos, controller)
    
    # Update Map
    map_global = localNav(abs_pos, goal_position, prox_horizontal, map_global)
    
    # Global Nav
    # goal_position = globalNav(map_global, abs_pos)
      
    # Update Motors
    
    
    # Thymio
    theta = abs_pos[0][2] - math.pi/2
    rotation_matrix = np.array([[np.cos(theta), np.sin(theta)],
                               [-np.sin(theta), np.cos(theta)]])
    
    thymio_coords_rotated = np.dot(thymio_coords, rotation_matrix)
    
    thymio_X = abs_pos[0][0]+thymio_coords_rotated[:,0]
    thymio_y = abs_pos[0][1]+thymio_coords_rotated[:,1]
    
    # Create a boolean mask for Thymio's shape in the global map
    thymio_mask = np.zeros_like(map_global, dtype=bool)
    thymio_X_grid = np.round(thymio_X).astype(int)
    thymio_y_grid = np.round(thymio_y).astype(int)
    thymio_mask[thymio_X_grid, thymio_y_grid] = True

    # Combine the global map with the Thymio's shape mask
    combined_map = np.logical_or(map_global, thymio_mask)

    # Find the bounding box of the free space around Thymio's shape
    free_space_indices = np.argwhere(combined_map == 0)
    min_x, min_y = np.min(free_space_indices, axis=0)
    max_x, max_y = np.max(free_space_indices, axis=0)
    
    # NMPC
    # Save the original stdout
    original_stdout = sys.stdout
    # Redirect stdout to a file or null device
    sys.stdout = open('output.log', 'w')
    
    controller_r, controller_l, states, time_points, controller = nmpc(abs_pos, goal_position, min_x, min_y, max_x, max_y)
    
    
    # Restore the original stdout
    sys.stdout = original_stdout

    #print(controller_r)
    #print(controller_l)
    
    #print(abs_pos[0][2]/3.1415)
    
    if math.sqrt( (abs_pos[0][0]-goal_position[0][0])**2 + (abs_pos[0][1]-goal_position[0][1])**2) > 4:
        v = {
        "motor.left.target": [int(controller_r) * 25],
        "motor.right.target": [int(controller_l) * 25],
        }
    else:
        v = {
        "motor.left.target": [0],
        "motor.right.target": [0],
        }
    
    aw(node.set_variables(v))
    
    
    # Plot results
    
    plt.figure(figsize=(10, 6))

    plt.subplot(4, 1, 1)
    plt.plot(time_points, states['x'], label='x')
    plt.xlabel('Time')
    plt.ylabel('Position x')
    plt.legend()

    plt.subplot(4, 1, 2)
    plt.plot(time_points, states['y'], label='y')
    plt.xlabel('Time')
    plt.ylabel('Position y')
    plt.legend()

    plt.subplot(4, 1, 3)
    plt.plot(time_points, states['theta'], label='theta')
    plt.xlabel('Time')
    plt.ylabel('Angle theta')
    plt.legend()

    plt.subplot(4, 1, 4)
    plt.plot(time_points, controller['u_r'], label='wr')
    plt.plot(time_points, controller['u_l'], label='wl')
    plt.xlabel('Time')
    plt.ylabel('Controller u')
    plt.legend()

    plt.tight_layout()
    plt.show()
    
    # Plot
    plot(abs_pos, thymio_coords_rotated, goal_position, map_global, abs_pos[0][2], min_x, min_y, max_x, max_y)
    plt.show()
    #time.sleep(0.1)
    clear_output(wait=True)  # Clear the previous plot
    
    #i = i+1
    
```
